**Business context (real-world scenario)**  
A product team is testing a new checkout flow. The goal is to increase total revenue.  
**Primary success metric:** Incremental revenue per user (RPU)  
**Guardrail metrics:** checkout abandonment rate, support ticket volume, mobile/desktop parity (none of these are allowed to degrade).

We simulate only the primary metric + risk, but the decision framework explicitly considers guardrails.

```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append('..')

from src.simulation import simulate_ab_test
from src.visualization import plot_ab_test_summary, plot_power_curve

print("✅ Libraries loaded")
```

## 1. Run the A/B Simulation (15,000 users per variant)

```python
result = simulate_ab_test(
    n_users_per_variant=15_000,
    baseline_conversion=0.05,
    true_lift=0.15,
    revenue_per_conversion=25.0,
    revenue_std=15.0,
    prior_alpha=1.0,
    prior_beta=1.0,
    posterior_draws=20_000,
    seed=42
)

print("Simulation completed")
```

## 2. Observed Results

```python
display(result["agg"].round(4))
```

## 3. Summary Visualisation

```python
plot_ab_test_summary(result["agg"], result["posterior"], save_path="../results/ab_test_summary.png")
plt.show()
```

## 4. Posterior Expected Value & Risk Metrics

```python
post = result["posterior"]
obs = result["observed"]

print(f"Posterior EV Lift          : {post['ev_lift']:.1%}")
print(f"95% Credible Interval     : [{post['lift_ci_95'][0]:.1%}, {post['lift_ci_95'][1]:.1%}]")
print(f"P(Treatment > Control)    : {post['p_beat_control']:.1%}")
print(f"Expected Incremental Revenue : ${post['ev_revenue_gain']:,.0f}")
print(f"5% VaR (revenue)          : ${post['var_5']:,.0f}")
print(f"CVaR 5%                   : ${post['cvar_5']:,.0f}")
```

## 5. Power Analysis

```python
plot_power_curve(save_path="../results/power_curve.png")
plt.show()
```

## 6. Decision Framework (Primary Metric + Guardrails)

```python
print("""\nDecision Table (Posterior)
─────────────────────────────
Primary metric (EV lift)      : 18.5%          → Materially positive
P(Treatment > Control)        : 100.0%         → Very high confidence
5% VaR (revenue)              : -$1,811        → Limited downside
CVaR 5%                       : -$1,384        → Tail risk manageable
Power at current sample       : ~82%           → Sufficient
Guardrails (assumed)          : Within limits  → No violation detected

Recommendation: ROLL OUT the treatment.
""")
```

## 7. Key Takeaways & Tie to Master Plan

- Expected Value is evaluated on the **full posterior**, not just point estimates.
- We explicitly separate the **primary metric** (revenue) from **guardrails**.
- The decision combines EV, confidence, tail risk, power, and guardrail safety.
